#Discrete Fourier transform

給定$N$個 sample data，$x_0,\dots , x_{N-1}$，discrete Fourier transform 是

$$X_k = \sum_{n=0}^{N-1} x_n e^{-i2\pi k n/N},\quad k= 0, \dots N-1.$$


首先我們引入需要的相關模組

```python
import numpy as np
import random
from matplotlib import pyplot as plt
import scipy
%matplotlib inline
```

底下我們產生一組週期sine wave，使用numpy提供的fft轉換後的結果

```python
L = 64
mySampleData = [np.sin(np.pi*a/L) for a in range(0, 1024)]
#myFreqRes = abs(scipy.fft(mySampleData))
myFreqRes = np.fft.fft(mySampleData)
print "Peak Value", myFreqRes[1024/2/L],myFreqRes[1024-1024/2/L]

plt.subplot(2, 1, 1)
plt.plot(mySampleData)
plt.title("Generated Sine Wave")

plt.subplot(2, 1, 2)
plt.plot(abs(myFreqRes))
plt.title("Generated Wave Spectrum")

plt.show()
```

上面結果的可以看到頻譜是對稱的，這是因為我們的訊號是real (參考[The real-input DFT-wiki](http://en.wikipedia.org/wiki/Discrete_Fourier_transform#The_real-input_DFT))，這個性質可以很容易的由DFT的定義得到。同時頻譜基本上是fourier series，所以可以知道理論上只會有兩項係數不為零，1024/(2L)以及1024-1024/(2L)


接著我們改取 L = 10，由於訊號總長為1024，所以

```python
L = 10
mySampleData = [np.sin(np.pi*a/L) for a in range(0, 1024)]
#myFreqRes = abs(scipy.fft(mySampleData))
myFreqRes = np.fft.fft(mySampleData)
print "1024/2/L=", 1024./2/L, "in INT =", 1024/2/L
print "Peak Value", myFreqRes[1024/2/L-2:1024/2/L+2]

plt.subplot(2, 1, 1)
plt.plot(mySampleData)
plt.title("Generated Sine Wave")

plt.subplot(2, 1, 2)
plt.plot(abs(myFreqRes))
plt.title("Generated Wave Spectrum")

plt.show()
```

和上個例子最大的差異是 由於訊號總數是1024，但是我們的wave是$\sin{\pi a/L}$，所以不是一個完整週期的sin/cos 波

#Inverse fast Fourier transform (IFFT)

底下我們使用numpy的ifft將結果賺換回時域

```python
L = 100
mySampleData = [np.sin(np.pi*a/L) for a in range(0, 1024)]
#myFreqRes = abs(scipy.fft(mySampleData))
myFreqRes = np.fft.fft(mySampleData)

plt.subplot(3, 1, 1)
plt.plot(mySampleData)
plt.title("Generated Sine Wave")

plt.subplot(3, 1, 2)
plt.plot(abs(myFreqRes))
plt.title("Generated Wave Spectrum")

plt.subplot(3, 1, 3)
plt.plot(np.fft.ifft(myFreqRes))
plt.title("Reconstructed Wave")
plt.show()
```

#線性疊加

```python
L1 = 64
mySampleData1 = [np.sin(np.pi*a/L1) for a in range(0, 1024)]
L2 = 32
mySampleData2 = [np.sin(np.pi*a/L2) for a in range(0, 1024)]
mySampleData = [sum(xx) for xx in zip(mySampleData1,mySampleData2)]
#myFreqRes = abs(scipy.fft(mySampleData))
myFreqRes = np.fft.fft(mySampleData)
print "Peak Value 1:", myFreqRes[1024/2/L1],myFreqRes[1024-1024/2/L1]
print "Peak Value 2:", myFreqRes[1024/2/L2],myFreqRes[1024-1024/2/L2]
plt.subplot(2, 1, 1)
plt.plot(mySampleData)
plt.title("Generated Sine Wave")

plt.subplot(2, 1, 2)
plt.plot(abs(myFreqRes))
plt.title("Generated Wave Spectrum")

plt.show()
```

上面的例子很清楚的告訴我們，兩個波線性疊加後，頻譜的變化。另外注意我們使用zip 把兩個list結合起來，再逐項加成為新的訊號。

#Denoise
去雜訊有非常多種技巧，同時不同種類的雜訊也有各自最適合的去雜訊技巧。在此我們考慮上面的訊號，觀察後發現雜訊在頻譜上的分量和主訊號差距甚大所以我們可以設定一個threshold，低於threshold直接設成0。底下我們可以看看效果如何

```python
L = 64
mySampleData = [np.sin(np.pi*a/L) for a in range(0, 1024)]
noise = np.random.normal(0,0.5,1024)
myData = [sum(xx) for xx in zip(mySampleData,noise)]
myFreqRes = np.fft.fft(myData)

maxval = abs(myFreqRes).max()
threshold = 0.1
#print maxval
tmp_array = []

for xx in myFreqRes:
    if abs(xx)> threshold * maxval:
        tmp_array.append(xx)
    else:
        tmp_array.append(0)

FilteredFreq = np.asarray(tmp_array)

FilteredData = np.fft.ifft(FilteredFreq)
plt.subplot(2, 2, 1)
plt.plot(myData)
plt.title("Generated Noise+Sine Wave")

plt.subplot(2, 2, 2)
#pp.plot(abs(scipy.fft(data)))
plt.plot(abs(myFreqRes))
plt.title("Generated Wave Spectrum")

plt.subplot(2, 2, 3)
plt.plot(FilteredData)
plt.title("Filtered Wave")

plt.subplot(2, 2, 4)
#pp.plot(abs(scipy.fft(data)))
plt.plot(abs(FilteredFreq))
plt.title("Filtered Wave Spectrum")

plt.show()
```

底下是[Simulating Noise in Python](http://stuartowen.com/home/2012/06/09/simulating-noise-in-python/#!/single_blog)文中設計的noise函數，目的在產生一個接近特定頻率的雜訊，大家可以分析看看

```python
class firFilter:

    coefficients = np.array([])

    stateVariables = np.array([])
    def __init__(self, filterCoeffs):
        if (type(filterCoeffs) == np.ndarray):
            # Save input filter coefficients
            self.coefficients = filterCoeffs
            # Initialize filter state variables
            self.stateVariables = np.array([0]*len(filterCoeffs))
        else:
            raise TypeError("Wrong type for filterCoeffs, "+
                "should be numpy.ndarray.")
            
    def processSample(self, inputSample):
        # Add new sample to state variables, remove oldest one
        self.stateVariables = np.append(np.delete(self.stateVariables, 0),
                                           inputSample)
        # Calculate filter output
        return sum(np.multiply(self.coefficients, self.stateVariables))
class shaper:

    # This list of doubles controls signal amplitude across the frequency
    # spectrum.
    sampleNoise =np.array( [])

    # This is te FIR filter instance used to filter noise
    noiseFilter = None
    def __init__(self, sampleData):
        # Check sampleData is of correct type
        if (type(sampleData) in [list, np.ndarray]):
            self.sampleNoise = np.array(sampleData)
        else:
            raise TypeError("Incorrect type for sample data"+
                        ".  It should be numpy.ndarray or list.")
        self.noiseFilter = firFilter(self.sampleNoise)
    def getNoise(self):
        return self.noiseFilter.processSample(random.random())
```

```python
mySampleData = [np.sin(np.pi*a/10) for a in range(0, 1024)]

x = shaper(sampleData = mySampleData)

for i in range(0, 5):
    data = []
    for i in range(0, 1024):
        data.append(x.getNoise())

    plt.subplot(2, 1, 1)
    plt.plot(data)
    plt.title("Generated Noise")

    plt.subplot(2, 1, 2)
    plt.plot(abs(scipy.fft(data)))
    plt.title("Generated Noise Spectrum")

plt.show()
```
