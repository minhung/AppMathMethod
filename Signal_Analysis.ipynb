```python
import numpy as np
import random
from matplotlib import pyplot as plt
import scipy
%matplotlib inline
```

```python
class firFilter:

    coefficients = np.array([])

    stateVariables = np.array([])
    def __init__(self, filterCoeffs):
        if (type(filterCoeffs) == np.ndarray):
            # Save input filter coefficients
            self.coefficients = filterCoeffs
            # Initialize filter state variables
            self.stateVariables = np.array([0]*len(filterCoeffs))
        else:
            raise TypeError("Wrong type for filterCoeffs, "+
                "should be numpy.ndarray.")
            
    def processSample(self, inputSample):
        # Add new sample to state variables, remove oldest one
        self.stateVariables = np.append(np.delete(self.stateVariables, 0),
                                           inputSample)
        # Calculate filter output
        return sum(np.multiply(self.coefficients, self.stateVariables))
```

```python
class shaper:

    # This list of doubles controls signal amplitude across the frequency
    # spectrum.
    sampleNoise =np.array( [])

    # This is te FIR filter instance used to filter noise
    noiseFilter = None
    def __init__(self, sampleData):
        # Check sampleData is of correct type
        if (type(sampleData) in [list, np.ndarray]):
            self.sampleNoise = np.array(sampleData)
        else:
            raise TypeError("Incorrect type for sample data"+
                        ".  It should be numpy.ndarray or list.")
        self.noiseFilter = firFilter(self.sampleNoise)
    def getNoise(self):
        return self.noiseFilter.processSample(random.random())
```

```python
mySampleData = [np.sin(np.pi*a/10) for a in range(0, 1024)]

x = shaper(sampleData = mySampleData)

noise = np.random.normal(0,1,1024)

for i in range(0, 5):
    data = []
for i in range(0, 1024):
    data.append(x.getNoise())

plt.subplot(2, 1, 1)
#pp.plot(data)
plt.plot(mySampleData)
plt.title("Generated Noise")

plt.subplot(2, 1, 2)
#pp.plot(abs(scipy.fft(data)))
plt.plot(abs(scipy.fft(mySampleData)))
plt.title("Generated Noise Spectrum")

plt.show()
```
